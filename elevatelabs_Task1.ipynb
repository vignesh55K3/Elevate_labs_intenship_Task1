## Handing SCHEDULEDDAY & APPOINTMENTDAY  columns

```python

from datetime import datetime as dt
data['SCHEDULEDDAY'] = pd.to_datetime(data['SCHEDULEDDAY'])
data["APPOINTMENTDAY"] = pd.to_datetime(data["APPOINTMENTDAY"])
# add a column for Year
```

```python
data
```

```python

data['WAITING_DAYS'] = (data['APPOINTMENTDAY'] - data['SCHEDULEDDAY']).dt.days

```

```python
data.WAITING_DAYS.max()
```

```python
data.WAITING_DAYS.min()
```

```python
neg_waiting_dates = data.loc[data['WAITING_DAYS']<0]
neg_waiting_dates
```

```python
data['WAITING_DAYS'].value_counts().sort_index()
```

```python
# Replace -1 with 0 
data['WAITING_DAYS'] = data['WAITING_DAYS'].replace(-1, 0)
# removes the negative value(eg: -7-2)
data = data[data['WAITING_DAYS'] >= 0]

```

```python
neg_waiting_dates1 = data.loc[data['WAITING_DAYS']<0]
neg_waiting_dates1
```

- removed all negative values 
- -1 is converted into 0
